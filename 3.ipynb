# Part 1

```python
example = """987654321111111
811111111111119
234234234234278
818181911112111"""
```

```python
def greedy(digits: str, desired_length: int) -> int:
    stack : list[str] = []
    to_remove = len(digits) - desired_length
    for digit in digits:
        while to_remove and stack and stack[-1] < digit:
            stack.pop()
            to_remove -= 1
        stack.append(digit)
    return int(''.join(stack[:desired_length]))

def solve_a(input_data: str) -> int:
  """Solves part A of the puzzle."""
  return sum(greedy(digits, 2) for digits in input_data.strip().split())
```

```python
assert(solve_a(example) == 357)
```

```python
puzzle.answer_a = solve_a(puzzle.input_data)
```

```python
def solve_b(input_data: str) -> int:
  """Solves part B of the puzzle."""
  return sum(greedy(digits, 12) for digits in input_data.strip().split())
```

```python
assert(solve_b(example) == 3121910778619)
```

```python
puzzle.answer_b = solve_b(puzzle.input_data)
```
